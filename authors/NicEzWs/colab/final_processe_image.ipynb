# Predict


```python
seg_model_path = '/content/data/model/best_seg.pt'
cls_model_path = '/content/data/model/best_angle_cls.pt'
image_path = '/content/data/raw_image/new/*'
```

```python
sr = 0.3
image_path_list = sorted(glob.glob(image_path))
image_list = [cv2.resize(cv2.imread(path), (0, 0), fx=sr, fy=sr) for path in image_path_list]
```

```python
model = YOLO(seg_model_path, 'segment')
result_list = model(image_path, imgsz=1024)
```

```python
!rm -r mask rotate cropped final_result
!mkdir mask rotate cropped final_result

# Process mask

for i, result in enumerate(result_list):
  index = i
  i = f"{i:03d}"

  masks = result.masks
  if masks == None:
    continue

  for j, mask_tensor in enumerate(masks.data):
    image = image_list[index]
    image_shape = image.shape

    label_index = result.boxes.cls.cpu().numpy().astype(np.uint8)[j].item()
    label_name = model.names[label_index]


    mask = mask_tensor.cpu().numpy().astype(np.uint8) * 255
    mask = cv2.resize(mask, (image_shape[1], image_shape[0]))

# Merge mask

    processed_img = cv2.bitwise_and(image, image, mask=mask)

    b, g, r = cv2.split(processed_img)
    image = cv2.merge([b, g, r, mask])

    cv2.imwrite(f'mask/mask_{i}_{j}.png', image)

# Rotate the image

    contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) == 0:
      raise ValueError('no contours found')

    (x, y), (w, h), angle = cv2.minAreaRect(contours[0])

    (h_img, w_img) = mask.shape

    side_length = h_img if h_img > w_img else w_img
    rect = (side_length, side_length)

    center = side_length // 2
    center = (center, center)

    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    image = cv2.warpAffine(image, M, rect, flags=cv2.INTER_LINEAR, borderValue=(0,0,0))
    mask = cv2.warpAffine(mask, M, rect, flags=cv2.INTER_NEAREST)

    cv2.imwrite(f'rotate/rotated_{i}_{j}.png', image)

# Crop the image

    ys, xs = np.where(mask == 255)

    if len(ys) > 0 and len(xs) > 0:
      x_min, x_max = xs.min(), xs.max()
      y_min, y_max = ys.min(), ys.max()

    image = image[y_min:y_max+1, x_min:x_max+1]
    mask = mask[y_min:y_max+1, x_min:x_max+1]

    cv2.imwrite(f'cropped/cropped_{i}_{j}.png', image)

# Rotate the image

    model = YOLO(cls_model_path)
    result_list = model(image)

    angle = int(model.names[result_list[0].probs.top1])
    angle = 90 - angle

    (h_img, w_img) = mask.shape

    side_length = h_img if h_img > w_img else w_img
    rect = (side_length, side_length)

    center = side_length // 2
    center = (center, center)

    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    image = cv2.warpAffine(image, M, rect, flags=cv2.INTER_LINEAR, borderValue=(0,0,0))
    mask = cv2.warpAffine(mask, M, rect, flags=cv2.INTER_NEAREST)

# Crop the image

    ys, xs = np.where(mask == 255)

    if len(ys) > 0 and len(xs) > 0:
      y_min, y_max = ys.min(), ys.max()
      x_min, x_max = xs.min(), xs.max()

    image = image[y_min:y_max+1, x_min:x_max+1]

    cv2.imwrite(f'final_result/result_{label_name}_{i}.png', image)
    cv2_imshow(image)
```
